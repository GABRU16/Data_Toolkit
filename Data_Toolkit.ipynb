1. What is NumPy, and why is it widely used in Python
- NumPy (Numerical Python) is a Python library used for numerical computing. It provides fast, efficient multidimensional arrays and mathematical functions for working with data.

- It is widely used because it is faster than Python lists, supports array operations, and is the foundation for libraries like Pandas, SciPy, and Matplotlib.

2.  How does broadcasting work in NumPy
- Broadcasting in NumPy allows arrays of different shapes to perform arithmetic operations by automatically expanding the smaller array to match the larger one, without copying data.

3.  What is a Pandas DataFrame
- A Pandas DataFrame is a two-dimensional table-like data structure with rows and columns. It is used to store, organize, and analyze data efficiently.

- It supports different data types in different columns and provides powerful functions for data manipulation and analysis.

4.  Explain the use of the groupby() method in Pandas.
- The groupby() method in Pandas is used to group data based on one or more columns and perform operations such as sum, count, mean, maximum, or minimum on each group.

- Example: Grouping sales data by category and calculating the total sales for each category.

5. Why is Seaborn preferred for statistical visualizations.
- Seaborn is preferred for statistical visualizations because it provides attractive, informative, and easy-to-create graphs. It has built-in support for statistical plots like histograms, box plots, violin plots, and heatmaps.

- It is built on Matplotlib, making it simpler to create high-quality visualizations with less code.

6. What are the differences between NumPy arrays and Python lists.
- NumPy arrays are faster and more memory-efficient than Python lists because they are designed for numerical computations. They store elements of the same data type and support vectorized operations, allowing mathematical calculations to be performed efficiently without loops. In contrast, Python lists can store different data types but are slower and require loops for most numerical operations. Therefore, NumPy arrays are preferred for scientific computing and data analysis.

7.  What is a heatmap, and when should it be used.
- A heatmap is a data visualization technique that uses colors to represent values in a matrix or dataset. Darker or brighter colors indicate higher or lower values, making patterns easy to see.

- It is used when we want to visualize relationships, correlations, or patterns in large datasets, such as correlation between variables or activity intensity in data.

8. What does the term “vectorized operation” mean in NumPy.
- A vectorized operation in NumPy means performing operations on entire arrays at once instead of using loops for each element.

- This is possible because NumPy applies operations internally in optimized C code, making it faster and more efficient than normal Python loops.

9. How does Matplotlib differ from Plotly.
- Matplotlib is a static plotting library used to create simple, highly customizable charts like line plots, bar charts, and histograms. It is widely used for basic data visualization in Python.

- Plotly is an interactive visualization library that creates dynamic charts where users can zoom, hover, and explore data. It is mainly used for dashboards and advanced interactive visualizations.

10. What is the significance of hierarchical indexing in Pandas.
- Hierarchical indexing in Pandas (also called MultiIndex) allows a DataFrame to have multiple levels of row or column labels. This helps represent and work with higher-dimensional data in a 2D table format.

- Its significance is that it enables more structured data organization, easier grouping, and advanced data analysis, especially when dealing with complex datasets like time series or grouped data.

11. What is the role of Seaborn’s pairplot() function.
- The pairplot() function in Seaborn is used to create a grid of plots that shows relationships between multiple numerical variables in a dataset.

- It plots scatter plots for every pair of variables and histograms or KDE plots on the diagonal, helping to quickly understand patterns, correlations, and distributions in the data.

12.  What is the purpose of the describe() function in Pandas.
- The describe() function in Pandas is used to generate a summary of statistical information for a DataFrame or Series.

- It provides key statistics like mean, median (50%), standard deviation, minimum, maximum, and quartiles (25%, 50%, 75%) for numerical data.

- It is mainly used for quick exploratory data analysis (EDA) to understand the distribution and central tendency of data.

13. Why is handling missing data important in Pandas.
- Handling missing data in Pandas is important because missing values can lead to incorrect analysis, biased results, and errors in calculations.

- If not handled properly, functions like mean, sum, or machine learning models may give wrong or incomplete outputs. Therefore, we use methods like fillna(), dropna(), or interpolation to clean the data and ensure accurate analysis.

14. What are the benefits of using Plotly for data visualization.
- The describe() function in Pandas provides a statistical summary of a dataset. It helps in quick exploratory data analysis (EDA) by showing key metrics such as count, mean, standard deviation, minimum, maximum, and quartile values (25%, 50%, 75%).

- It gives a clear overview of the distribution and spread of data, helping to understand the dataset before applying further analysis or machine learning.

15.  How does NumPy handle multidimensional arrays.
- NumPy handles multidimensional arrays using the ndarray (N-dimensional array) object. It can store data in multiple dimensions such as 1D (vectors), 2D (matrices), and higher-dimensional arrays (tensors).

- It supports efficient operations like indexing, slicing, reshaping, and mathematical computations across all dimensions, making it very useful for scientific and numerical computing.

16.  What is the role of Bokeh in data visualization.
- Bokeh is a Python library used for creating interactive and real-time data visualizations in web browsers.

- Its main role is to help build dynamic dashboards and visual plots where users can zoom, pan, and hover over data. It is especially useful for large datasets and web-based applications, making data exploration more interactive and user-friendly.

17. Explain the difference between apply() and map() in Pandas.
- In Pandas, both apply() and map() are used to transform data, but they work differently.

- map() is used only on a Series and is mainly for element-wise transformations, often with dictionaries, functions, or lambda expressions.
apply() is more flexible and can be used on both Series and DataFrames to perform operations along rows or columns.

18.  What are some advanced features of NumPy.
- Some advanced features of NumPy include broadcasting, vectorized operations, and multidimensional array support (ndarrays). It also provides powerful tools like linear algebra functions, Fourier transforms, random number generation, and array reshaping/slicing.

- These features make NumPy highly efficient for scientific computing, data analysis, and machine learning tasks.

19.  How does Pandas simplify time series analysis.
- Pandas simplifies time series analysis by providing powerful tools for working with date and time data. It allows easy date parsing, indexing with DatetimeIndex, resampling (daily, monthly, yearly), and time-based grouping.

- It also supports operations like shifting, rolling windows, and handling missing time values, making it much easier to analyze trends and patterns over time.

20. What is the role of a pivot table in Pandas.
- A pivot table in Pandas is used to summarize and reorganize data in a DataFrame. It helps in converting long data into a structured table format based on index, columns, and values.

- Its main role is to perform aggregation (like sum, mean, count) and make it easier to analyze patterns and relationships in data. It is widely used in data analysis and reporting.

21. Why is NumPy’s array slicing faster than Python’s list slicing.
- NumPy array slicing is faster because NumPy arrays are stored in contiguous memory blocks and use a fixed data type, allowing efficient access without extra overhead.

- In contrast, Python lists store references to objects, which adds extra memory and processing overhead. Also, NumPy performs slicing at a low-level optimized C implementation, making it much faster than Python list slicing.

22. What are some common use cases for Seaborn?
- Seaborn is commonly used for statistical data visualization. Some key use cases include creating distribution plots (histograms, KDE plots), relationship plots (scatter plots, pair plots), and categorical plots (box plots, bar plots, violin plots).

- It is also widely used for correlation heatmaps and exploratory data analysis (EDA) because it makes complex visualizations simple and visually appealing with less code.
